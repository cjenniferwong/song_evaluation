```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import pickle
```

```python
file_name = 'data/genres_encoded.pkl'
artist_data = pickle.load(open(file_name, 'rb'))
artist_data.head()
```

```python
genre_columns = [column for column in artist_data.columns if 'genre_' in column]
genre_columns
```

```python
genre_subset = artist_data[genre_columns].copy(deep=True)
genre_subset.sum().sort_values(ascending=False)
```
