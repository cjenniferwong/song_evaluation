````python
with open('agg_functions.json') as filename:
    agg_funcs = json.load(filename)
    filename.close()

def find_genres(artist_names):
    '''
    the artists column is a strings so i need to do this for every artist in the list
    '''
    artists_list = [artist_names.split(', ')]
    df_list = []
    for artist in artists_list:
        queries = f'name == {artist}'
        df = genres.query(queries)
#         print(df.columns)
        df.drop(['id', 'query', 'name'], axis=1, inplace=True)
        df_list.append(df)
    return pd.concat(df_list, axis=0) # essentially appends each df to each other
    
def get_genres(artist_names):
    '''
    takes the df returned from the find_genres and manup
    '''
    df = find_genres(artist_names)
    
    return df.agg(agg_funcs)
````

````python
find_genres(with_artists['artists'].values[0]).agg(agg_funcs)
````

````python
got_genres = with_artists['artists'].apply(get_genres)
got_genres
````
