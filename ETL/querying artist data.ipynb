```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import pickle
```

```python
file_path = 'data/artist_query_results.pkl'
artist_data = pickle.load(open(file_path, 'rb'))
artist_data.head(20)
```

```python
artist_data.info()
```

```python
# to clean it even more i need to mapp the genres into buckets possible?
genres_df = pd.get_dummies(artist_data['genres'].apply(pd.Series).stack(), prefix='genre').sum(level=0)
genres_df
```

```python
genres_encoded = pd.merge(artist_data.drop('genres', axis=1), genres_df, left_index=True, right_index=True, how ='left').fillna(0)
genres_encoded.head(20)
```

```python
genres_encoded.rename(columns={'followers':'artist_followers_count',
                              'popularity': 'aritst_popularity_score'}, inplace=True)
genres_encoded.head(20)
```

```python
pickle.dump(genres_encoded, open('data/genres_encoded.pkl','wb'))
#note to self, i should go through to check how well the code worked
# also, is there signal in songs that are labelled with genres? something to consider later on when feature engineering
```
