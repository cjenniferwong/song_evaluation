## Transforming the Data from querying the audio_features end point

```python
import pandas as pd

import pickle
```

```python
file_name = '../data/audio_features.pkl'

with open(file_name, 'rb') as file:
    audio_features = pickle.load(file)
    audio_features = pd.DataFrame.from_records(audio_features)
    file.close()
```

```python
audio_features.head(10)
```

```python
audio_features.columns
```

```python
# so now i need to figure out what features I want and which ones that I dont want
remove_columns = ['analysis_url', 'track_href']

audio_features.drop(remove_columns, axis=1, inplace=True, errors='ignore')
audio_features.head()
```

```python
set(audio_features['type'].values) # i should drop the type too since it doesn't provide signal for me
```

```python
audio_features.drop('type', axis=1, inplace=True, errors='ignore')
audio_features.head()
```

```python
# saving it

pickle.dump(audio_features, open('../data/audio_features.pkl', 'wb'))
```
