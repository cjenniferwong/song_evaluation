```python
import pandas as pd

import pickle
```

```python
spotify_results = pickle.load(open('../data/spotify_api_search_results.pkl', 'rb'))
spotify_results.head()

missing = spotify_results[spotify_results['name'].isna()]
missing.head()
```

```python
spotify_results.tail()
```

```python
spotify_results.info()
```

```python
missing_filled = pd.read_csv('../data/missing_modified_results.csv')
missing_filled.head()
# missing['search_query'] = missing['artists']
missing_filled['search_query'] = missing['search_query'].values
missing_filled.head()
```

```python
billboard_data = pd.read_csv('../data/billboards.csv')
billboard_data.head()
```

```python
billboard_data['search_query'] = billboard_data['Artist'].str.lower() + ' ' + billboard_data['Song'].str.lower()
billboard_data.head()
```

```python
num_weeks = pd.merge(spotify_results.dropna(), billboard_data[['search_query', 'num_weeks']],
                     left_on='search_query', right_on='search_query')
num_weeks.head()
```

```python
missing_num_weeks = pd.merge(missing_filled.dropna(), billboard_data[['search_query', 'num_weeks']],
                     left_on='search_query', right_on='search_query', how='left')
missing_num_weeks
```

```python
pd.concat([num_weeks, missing_num_weeks], axis=0).to_csv('../data/num_weeks.csv', index=False)
```
